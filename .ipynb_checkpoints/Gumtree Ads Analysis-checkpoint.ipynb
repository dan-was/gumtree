```python
cols = ['price','advertiser','n_rooms','location','size_m2']
dataset = dataset[cols]
sns.heatmap(dataset.isnull(),cbar=False,yticklabels=False)
```

```python
dataset.dropna(inplace=True)
```

```python
location = pd.get_dummies(dataset['location'])
```

```python
advertiser = pd.get_dummies(dataset['advertiser'],drop_first=True)
```

```python
dataset = pd.concat([dataset,advertiser,location],axis=1)
dataset.drop(['location','advertiser'],axis=1,inplace=True)
```

```python
dataset.head()
```

## Linear regression model

```python
X = dataset.drop('price',axis=1)
y = dataset['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
coeffs = pd.DataFrame(data=lm.coef_,index=X.columns, columns=['Coefficient'])
```

```python
coeffs.sort_values(by='Coefficient',inplace=True)
```

```python
sns.barplot(data=coeffs.transpose(),orient='h',color='lightblue')
```

```python
pred = lm.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,pred)
```
